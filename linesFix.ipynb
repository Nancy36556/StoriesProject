```python
import re
def linesFix(inputFile, outputFile):
    def find_lowercase_lines(text):
        pattern = r"^[a-z]"
        lowercase_lines = []
        for line in text.split("\n"):
            if re.search(pattern, line):
                lowercase_lines.append(line)
        return lowercase_lines

    def fixDots(text):
        pattern = r"(?<!(.{1}[Mm]r|[Mm]rs|.{1}[Dd]r))\."
        

        matches = re.finditer(pattern, text)


        lines = []
        i = 0
        for match in matches:
            start = match.start()
            end = match.end()
            
            text = text[:end+i] + '\n' + text[end+i:]
            i += 1


            line_start = text.rfind("\n", 0, start) + 1
            line_end = text.find("\n", end)

            if line_end == -1:
                line_end = len(text)
                

            line = text[line_start:line_end]
            lines.append(line)

            print("match number ", i, 'from', len(matches))

    
        return lines    


    f = open(inputFile, 'r', encoding="utf-8")
    data = f.read()

    text = (data.strip())
    text = re.sub(r'\s{2,}', '\n', text)

    lowercase_lines = find_lowercase_lines(text)

    counter = 0 
    lenn = len(text)
    for i in lowercase_lines:

        text = text[:text.find(i)-1] + ' ' +text[text.find(i):lenn + counter -1]
        counter += 1

        print('line number ', counter, 'of ', len(lowercase_lines))




    with open(outputFile, "w", encoding="utf-8") as fOut:
        for i in fixDots(text):

            fOut.writelines(i + '\n')


    with open(outputFile, 'r', encoding="utf-8") as f:
        data = f.read()
        text = (data.strip())
        text = re.sub(r'\s{2,}', '\n', text) 

    with open(outputFile, "w", encoding="utf-8") as fOut:
        for i in fixDots(text):

            fOut.writelines(i + '\n')        
          
```
